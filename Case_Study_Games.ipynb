```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sheet_url='https://docs.google.com/spreadsheets/d/10poofg-l8DMdtUgGy8mOpra2IHmA9EQC7drmF9AyYHA/edit#gid=1485085913'
data = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df=pd.read_csv(data)

df.head()
```

```python
df.info()
```

```python
df_cleaned=df.copy()
df_cleaned['Release']=pd.to_datetime(df_cleaned['Release'])
df_cleaned=df_cleaned.fillna('no data')
df_cleaned.head()
```

```python
 df_cleaned.info()
```

```python
df_cleaned['Name'].duplicated().sum()
```

```python
df_duplicate=df_cleaned[df_cleaned['Name'].duplicated()]
df_duplicate
```

Questions 1 :Which game is the oldest and the newest games in that dataset?

```python
#1.1 The oldest games
df_sorted=df_cleaned.sort_values(by='Release')
df_sorted.iloc[0]
```

```python
#1.2 The latest games
df_sorted.iloc[-1]
```

Questions 2: Which publisher published most of the games?

```python
#2. Publisher with the most games published
for publisher in ['Publisher']:
  print(df_cleaned[publisher].value_counts())
```

```python
pd.set_option('display.max_rows',None)
df_publisher=df_cleaned.groupby(['Publisher']).sum().reset_index()
df_publisher
```

Questions 3: Which developer developed most of the games?

```python
#3. Developer with the most games developed
for developer in ['Developer']:
  print(df_cleaned[developer].value_counts().head())
```

```python
pd.set_option('display.max_rows',None)
df_developer=df_cleaned.groupby(['Developer']).sum()
df_developer
```

Questions 4 : Which series is the most sales?

```python
#4. Series with the most sales
df_series_sales=df_cleaned.groupby(['Series']).sum().sort_values(by='Sales')
df_series_sales.tail()
```

Questions 5 : Which series have the most games?

```python
#5. Series with te most games
pd.set_option('display.max_rows',5)
for series in ['Series']:
  print(df_cleaned[series].value_counts())
```
